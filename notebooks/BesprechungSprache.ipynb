```python
import pandas as pd
```

```python
df = pd.read_csv('../dataset/movies_complete.csv').dropna(subset=['Language'])
```

```python
df.iloc[df.Language.apply(lambda x: len(x)).argmax()].Language
```

```python
num_langs = df.Language.str.split(', ').apply(lambda x: len(x))
```

```python
import numpy as np
num_langs.plot(kind='hist', bins=np.unique(num_langs))
```

```python
num_langs[num_langs > 1].shape[0]
```

```python
df[num_langs > 1].Language.value_counts().plot(kind='bar')
```

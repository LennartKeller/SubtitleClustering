# Merges our dataset with the imdb datasets to obtain additional metadata

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../dataset/subtitles_dataset_tokenized.csv')
crew_df = pd.read_csv('../dataset/crew.tsv', sep='\t')
title_df = pd.read_csv('../dataset/titles.csv', sep='\t')
ratings_df = pd.read_csv('../dataset/ratings.csv', sep='\t')
```

## Convert imdb ids from main dataset to original format (=> adding tt prefix)

```python
df['IMDB_ID'] = df['IMDB_ID'].apply(lambda x: f"tt{x}")
```

```python
df.head()
```

## Merge crew dataset into main dataset

```python
crew_df.head()
```

```python
crew_df.rename(columns={'tconst': 'IMDB_ID'}, inplace=True)
df = df.merge(right=crew_df, on='IMDB_ID', how='outer')
#df = df.drop(['directors_x', 'writers_x', 'directors_y', 'writers_y'], axis=1)
```

```python
df.head()
```

## Merge title dataset into main dataset

```python
title_df.rename(columns={'tconst': 'IMDB_ID'}, inplace=True)
df = df.merge(right=title_df, on='IMDB_ID', how='outer')
df = df.replace(r'\N', np.nan)
```

```python
df.head()
```

## Merge ratings dataset into main dataset

```python
ratings_df.rename(columns={'tconst': 'IMDB_ID'}, inplace=True)
df = df.merge(right=ratings_df, on='IMDB_ID', how='outer')
df = df.replace(r'\N', np.nan)
```

```python
df.head()
```

### Set index, convert emtpy string fields to nan, etc...

```python
df.replace(r'\N', np.nan, inplace=True)
df.set_index('filename', inplace=True)
df.drop('Unnamed: 0', axis=1, inplace=True)
df.drop('genres', axis=1, inplace=True)
df.year = df.year.replace(0, np.nan)
```

```python
df.shape[0], df.dropna().shape[0]
```

```python
df.head()
```

### Save dataset 

```python
df.to_csv('../dataset/complete_dataset.csv')
```

# Merge openIMDB data into movie_spacy dataset

```python
movies = pd.read_csv('../dataset/movies_spacy.csv')
openimdb = pd.read_csv('../dataset/openIMDB.csv')
```

```python
movies.columns
```

```python
openimdb.columns
```

```python
merged = movies.merge(openimdb, left_on='IMDB_ID', right_on='imdbID')
```

```python
merged.columns
```

```python
for column in merged.columns:
    if "Unnamed:" in column:
        merged = merged.drop(column, axis=1)
```

```python
merged.columns
```

```python
merged.to_csv('../dataset/movies_complete.csv')
```
